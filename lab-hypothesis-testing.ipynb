```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.spatial import distance
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df.rename(columns={col : col.strip() for col in df.columns}, inplace = True)
df.rename(columns={col : col.replace(" ", "_").lower() for col in df.columns}, inplace = True)
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# we set our significance level
alpha = 0.05

hp_dragon = df[df['type_1'] == 'Dragon']['hp']
hp_grass = df[df['type_1'] == 'Grass']['hp']

t_stat, p_value = st.ttest_ind(hp_dragon, hp_grass, equal_var=False, alternative='greater')

display(t_stat, p_value)
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# independent t-test

legendary = df[df['legendary'] == True]
non_legendary = df[df['legendary'] == False]

stats_list = ['hp', 'attack', 'defense', 'sp._atk', 'sp._def', 'speed']

alpha = 0.05

for stat in stats_list:
    legendary_groupped = legendary[stat].dropna()
    non_legendary_grupped = non_legendary[stat].dropna()

    t_stat, p_value = st.ttest_ind(legendary_groupped, non_legendary_grupped, equal_var=False)


    print(f"\n Stats: {stat}")
    print(f"   p-value: {p_value:.15f}")

    if p_value < 0.05:
        print("  We reject the H0 hypotesis.")
        print("  there is evidence that legendary and non-legendary pokemon have similar "+ stat +" ")
    else:
        print("  We do not reject the H0 hypotesis")
        print("  there is not enough evidence that the " + stat + " are different between both groups")
```

```python
# Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.

def distance_calculator(longitude, latitude):

    school_coord = (-114.40, 34.00)
    hospital_coord = (-114.50, 33.80)

    house_coord = (longitude, latitude)

    school_dist = distance.euclidean(house_coord, school_coord)
    hospital_dist = distance.euclidean(house_coord, hospital_coord)

    return school_dist, hospital_dist
```
